# 함수와 클래스

## 함수(Function)란?
- 특정한 작업을 수행하는 코드의 묶음
- 코드의 재사용성을 높이고 프로그램의 구조화를 위해 사용
- 입력값(매개변수)을 받아서 처리한 후 결과값(반환값)을 돌려줄 수 있음

# 함수의 기본 구조

## 함수의 생김새
```python
def 함수이름(매개변수1, 매개변수2, ...):
    실행할 코드
    return 반환값
```

## 함수의 구성요소
- def: 함수를 정의하는 키워드
- 함수이름: 함수를 호출할 때 사용할 이름
- 매개변수: 함수에 전달되는 입력값 (선택사항)
- return: 함수의 결과값을 반환 (선택사항)

````python
# 함수 예제
def 인사하기():
    print("안녕하세요!")

def 더하기(a, b):
    return a + b

# 함수 호출하기
인사하기()
결과 = 더하기(3, 5)
print("3 + 5 =", 결과)
````

# 클래스(Class)란?
- 객체를 만들기 위한 설계도
- 데이터(속성)와 이를 처리하는 메서드(함수)를 하나로 묶어서 관리
- 객체 지향 프로그래밍의 기본 단위

## 클래스의 생김새
```python
class 클래스이름:
    def __init__(self, 매개변수1, 매개변수2, ...):
        self.속성1 = 매개변수1
        self.속성2 = 매개변수2
        
    def 메서드1(self):
        실행할 코드
        
    def 메서드2(self):
        실행할 코드
```

## 클래스의 구성요소
- class: 클래스를 정의하는 키워드
- __init__: 클래스의 생성자 (객체가 생성될 때 자동으로 호출)
- self: 객체 자신을 가리키는 참조
- 속성(attribute): 객체의 데이터
- 메서드(method): 객체의 동작을 정의하는 함수


# 객체 지향 프로그래밍 (OOP)

## 객체(Object)란?
- 실제 세계의 물체나 개념을 프로그래밍으로 표현한 것
- 데이터(속성)와 기능(메서드)을 가지고 있는 프로그램의 기본 단위
- 예시: 
  - 자동차 객체 → 속성(색상, 브랜드, 속도), 기능(시동걸기, 가속하기, 정지하기)
  - 학생 객체 → 속성(이름, 학번, 학년), 기능(공부하기, 시험보기, 출석하기)

## 객체 지향 프로그래밍의 특징
1. 캡슐화 (Encapsulation)
   - 데이터와 기능을 하나로 묶어서 관리
   - 정보 은닉을 통해 객체의 내부 구현을 숨김
   - 예: private 변수, getter/setter 메서드

2. 상속 (Inheritance)
   - 기존 클래스의 특성을 다른 클래스가 물려받음
   - 코드 재사용성을 높이고 계층 구조를 만듦
   - 예: 동물 클래스 → 강아지 클래스, 고양이 클래스

3. 다형성 (Polymorphism)
   - 같은 이름의 메서드가 다른 기능을 할 수 있음
   - 오버라이딩: 부모 클래스의 메서드를 자식 클래스에서 재정의
   - 예: 동물.소리내기() → 강아지는 "멍멍", 고양이는 "야옹"

4. 추상화 (Abstraction)
   - 공통적인 특성을 뽑아내어 상위 개념을 만듦
   - 복잡한 내용은 숨기고 핵심적인 내용만 표현
   - 예: 자동차의 시동걸기 → 내부 엔진 작동 과정은 숨김


````python
# OOP 예제
class 자동차:
    def __init__(self, 브랜드, 색상):
        # 공개 속성
        self.브랜드 = 브랜드
        # 비공개 속성 (캡슐화)
        self.__속도 = 0
        self.__색상 = 색상
        self.__시동 = False
    
    # getter 메서드
    def 현재속도(self):
        return self.__속도
    
    # setter 메서드
    def 색상변경(self, 새색상):
        self.__색상 = 새색상
        print(f"색상이 {새색상}으로 변경되었습니다.")
    
    def 시동걸기(self):
        if not self.__시동:
            self.__시동 = True
            print("시동이 걸렸습니다.")
        else:
            print("이미 시동이 걸려있습니다.")
    
    def 가속(self, 속도증가):
        if self.__시동:
            self.__속도 += 속도증가
            print(f"현재 속도: {self.__속도}km/h")
        else:
            print("먼저 시동을 걸어주세요.")

# 객체 생성 및 사용
내차 = 자동차("현대", "검정")
내차.시동걸기()
내차.가속(30)
내차.색상변경("빨강")
print(f"현재 속도는 {내차.현재속도()}km/h 입니다.")
````

````python
# 일반적인 상속 예제
class 동물:
    def __init__(self, 이름):
        self.이름 = 이름
    
    def 소리내기(self):
        print("동물이 소리를 냅니다")
    
    def 먹기(self):
        print(f"{self.이름}이(가) 먹이를 먹습니다")

# 동물 클래스를 상속받는 강아지 클래스
class 강아지(동물):
    def 소리내기(self):  # 부모 메서드를 재정의 (오버라이딩)
        print("멍멍!")
    
    def 산책하기(self):  # 새로운 메서드 추가
        print(f"{self.이름}이(가) 산책을 합니다")

# 동물 클래스를 상속받는 고양이 클래스
class 고양이(동물):
    def 소리내기(self):  # 부모 메서드를 재정의 (오버라이딩)
        print("야옹!")

print("=== 일반 상속 예제 ===")
멍멍이 = 강아지("멍멍이")
야옹이 = 고양이("야옹이")

멍멍이.소리내기()  # 자식 클래스에서 재정의한 메서드 사용
멍멍이.먹기()      # 부모 클래스의 메서드 그대로 사용
멍멍이.산책하기()  # 자식 클래스에서 추가한 메서드 사용

야옹이.소리내기()  # 자식 클래스에서 재정의한 메서드 사용
야옹이.먹기()      # 부모 클래스의 메서드 그대로 사용
````

````python
# 추상 클래스 예제
from abc import ABC, abstractmethod

class 도형(ABC):  # ABC를 상속받아 추상 클래스 생성
    @abstractmethod  # 추상 메서드 표시
    def 넓이(self):
        pass  # 구현은 자식 클래스에서 반드시 해야 함
    
    @abstractmethod
    def 그리기(self):
        pass

# 추상 클래스를 상속받는 구체 클래스
class 사각형(도형):
    def __init__(self, 가로, 세로):
        self.가로 = 가로
        self.세로 = 세로
    
    def 넓이(self):  # 반드시 구현해야 함
        return self.가로 * self.세로
    
    def 그리기(self):  # 반드시 구현해야 함
        print(f"가로 {self.가로}, 세로 {self.세로}인 사각형을 그립니다")

class 원(도형):
    def __init__(self, 반지름):
        self.반지름 = 반지름
    
    def 넓이(self):  # 반드시 구현해야 함
        return 3.14 * self.반지름 ** 2
    
    def 그리기(self):  # 반드시 구현해야 함
        print(f"반지름 {self.반지름}인 원을 그립니다")

print("\n=== 추상 클래스 예제 ===")
# 도형1 = 도형()  # 오류! 추상 클래스는 인스턴스를 직접 만들 수 없음

사각형1 = 사각형(가로=5, 세로=3)
원1 = 원(반지름=4)

사각형1.그리기()
print(f"사각형의 넓이: {사각형1.넓이()}")

원1.그리기()
print(f"원의 넓이: {원1.넓이()}")

# 추상 클래스의 메서드를 구현하지 않으면?
class 삼각형(도형):  # 오류! 추상 메서드를 모두 구현하지 않았음
    def __init__(self, 밑변, 높이):
        self.밑변 = 밑변
        self.높이 = 높이
    
    def 넓이(self):  # 하나만 구현하고 그리기()는 구현하지 않음
        return self.밑변 * self.높이 / 2

# 삼각형1 = 삼각형(4, 3)  # 오류! 추상 메서드를 모두 구현하지 않아서 인스턴스를 만들 수 없음
````

# 파이썬 내장 객체의 메소드 사용하기

파이썬에서는 리스트, 튜플, 문자열과 같은 내장 객체들도 클래스로 만들어져 있어서 `.메소드()` 형식으로 다양한 기능을 사용할 수 있습니다.

## 1. 리스트(List) 메소드
- `.append(x)`: 리스트 끝에 항목 추가
- `.extend(iterable)`: 리스트 끝에 여러 항목 추가
- `.insert(i, x)`: 지정한 위치 i에 항목 x를 삽입
- `.remove(x)`: 값이 x인 첫 번째 항목 삭제
- `.pop([i])`: i 위치의 항목을 삭제하고 반환
- `.sort()`: 리스트를 정렬
- `.reverse()`: 리스트를 역순으로 뒤집기
- `.count(x)`: 값이 x인 항목의 개수를 반환
- `.index(x)`: 값이 x인 첫 번째 항목의 인덱스를 반환

## 2. 문자열(String) 메소드
- `.upper()`: 모든 문자를 대문자로 변환
- `.lower()`: 모든 문자를 소문자로 변환
- `.strip()`: 양쪽 공백 제거
- `.replace(old, new)`: old를 new로 치환
- `.split(sep)`: 구분자를 기준으로 문자열을 나누어 리스트로 반환
- `.join(iterable)`: 문자열을 구분자로 사용하여 여러 문자열을 하나로 결합
- `.startswith(prefix)`: 지정한 접두사로 시작하는지 확인
- `.endswith(suffix)`: 지정한 접미사로 끝나는지 확인

## 3. 세트(Set) 메소드
- `.add(x)`: 항목 추가
- `.remove(x)`: 항목 제거 (항목이 없으면 에러)
- `.discard(x)`: 항목 제거 (항목이 없어도 에러 없음)
- `.union(other)`: 합집합
- `.intersection(other)`: 교집합
- `.difference(other)`: 차집합

## 4. 딕셔너리(Dictionary) 메소드
- `.keys()`: 모든 키를 반환
- `.values()`: 모든 값을 반환
- `.items()`: 모든 (키, 값) 쌍을 반환
- `.get(key[, default])`: 키에 해당하는 값을 반환 (없으면 default 반환)
- `.update(other)`: 다른 딕셔너리의 내용을 추가
- `.pop(key[, default])`: 키에 해당하는 항목을 삭제하고 값을 반환


````python
# 파이썬 내장 객체들이 어떻게 구현되어 있는지 비슷하게 만들어보기

# 1. 리스트와 비슷한 클래스 만들기
class MyList:
    def __init__(self):
        self.data = []  # 실제 데이터를 저장할 리스트
    
    def append(self, item):
        self.data = self.data + [item]
        
    def extend(self, items):
        self.data = self.data + list(items)
    
    def insert(self, index, item):
        self.data = self.data[:index] + [item] + self.data[index:]
    
    def remove(self, item):
        new_data = []
        removed = False
        for x in self.data:
            if x == item and not removed:
                removed = True
                continue
            new_data.append(x)
        self.data = new_data
    
    def sort(self):
        # 간단한 버블 정렬 구현
        n = len(self.data)
        for i in range(n):
            for j in range(0, n-i-1):
                if self.data[j] > self.data[j+1]:
                    self.data[j], self.data[j+1] = self.data[j+1], self.data[j]
    
    def __str__(self):
        return str(self.data)

# 2. 문자열과 비슷한 클래스 만들기
class MyString:
    def __init__(self, text):
        self.text = str(text)
    
    def upper(self):
        result = ""
        for char in self.text:
            if 'a' <= char <= 'z':
                result += chr(ord(char) - 32)  # ASCII 코드 활용
            else:
                result += char
        return MyString(result)
    
    def lower(self):
        result = ""
        for char in self.text:
            if 'A' <= char <= 'Z':
                result += chr(ord(char) + 32)  # ASCII 코드 활용
            else:
                result += char
        return MyString(result)
    
    def strip(self):
        # 앞뒤 공백 제거
        start = 0
        end = len(self.text) - 1
        
        while start <= end and self.text[start] == ' ':
            start += 1
        while end >= start and self.text[end] == ' ':
            end -= 1
            
        return MyString(self.text[start:end+1])
    
    def split(self, sep=None):
        if sep is None:
            sep = ' '
        result = MyList()
        current = ""
        
        for char in self.text:
            if char == sep:
                if current:
                    result.append(current)
                current = ""
            else:
                current += char
        
        if current:
            result.append(current)
        
        return result
    
    def __str__(self):
        return self.text

# 3. 세트와 비슷한 클래스 만들기
class MySet:
    def __init__(self, items=None):
        self.data = []
        if items:
            for item in items:
                self.add(item)
    
    def add(self, item):
        if item not in self.data:
            self.data.append(item)
    
    def remove(self, item):
        if item in self.data:
            self.data.remove(item)
        else:
            raise KeyError(f"{item} not found in set")
    
    def union(self, other):
        result = MySet(self.data)
        for item in other.data:
            result.add(item)
        return result
    
    def intersection(self, other):
        result = MySet()
        for item in self.data:
            if item in other.data:
                result.add(item)
        return result
    
    def __str__(self):
        return "{" + ", ".join(str(x) for x in self.data) + "}"

# 사용 예제
print("=== MyList 테스트 ===")
my_list = MyList()
my_list.append(1)
my_list.append(3)
my_list.append(2)
print("원본:", my_list)
my_list.sort()
print("정렬 후:", my_list)

print("\n=== MyString 테스트 ===")
my_str = MyString("  Hello, World!  ")
print("원본:", my_str)
print("대문자:", my_str.upper())
print("소문자:", my_str.lower())
print("공백제거:", my_str.strip())
print("분리:", my_str.split(','))

print("\n=== MySet 테스트 ===")
set1 = MySet([1, 2, 3, 2])  # 중복은 자동으로 제거됨
set2 = MySet([2, 3, 4])
print("set1:", set1)
print("set2:", set2)
print("합집합:", set1.union(set2))
print("교집합:", set1.intersection(set2))
````

# 자동차를 함수로만 표현하면...

- 클래스 단위로 코드를 나누면 여러 사람이 각자 맡은 부분을 개발하기 쉬워요
- 다른 사람이 만든 클래스를 가져다 쓸 때도 사용법을 이해하기 쉬워요


````python
# 함수로만 만들 경우
user_name1 = "철수"
user_age1 = 20
user_scores1 = [90, 85, 95]

user_name2 = "영희"
user_age2 = 21
user_scores2 = [80, 95, 70]

def print_user_info(name, age, scores):
    print(f"이름: {name}, 나이: {age}, 점수: {scores}")

print_user_info(user_name1, user_age1, user_scores1)
print_user_info(user_name2, user_age2, user_scores2)
````

````python
# 클래스로 만들 경우
class User:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.scores = []
    
    def print_info(self):
        print(f"이름: {self.name}, 나이: {self.age}")
    
    def add_score(self, score):
        self.scores.append(score)

# 클래스 사용 예시
철수 = User("철수", 20)
영희 = User("영희", 21)

# 점수 추가
철수.add_score(90)
철수.add_score(85)
철수.add_score(95)

영희.add_score(80)
영희.add_score(95)
영희.add_score(70)

print(f"이름: {철수.name}, 나이: {철수.age}, 점수: {철수.scores}")
print(f"이름: {영희.name}, 나이: {영희.age}, 점수: {영희.scores}")
````

위 코드를 보면 클래스를 사용했을 때의 장점이 더 잘 드러나요:

1. **데이터 관리가 편해요**: 
   - 함수만 사용할 때는 각 학생의 정보(이름, 나이, 점수)를 따로따로 관리해야 했어요
   - 클래스를 사용하면 한 학생의 모든 정보가 하나의 객체에 깔끔하게 담겨요

2. **기능 추가가 쉬워요**:
   - `add_score` 메서드를 추가해서 점수를 관리하기 더 쉽게 만들었어요
   - 나중에 새로운 기능이 필요하면 클래스에 메서드만 추가하면 돼요

3. **사용하기 더 직관적이에요**:
   - 함수만 사용할 때: `print_user_info(user_name1, user_age1)`
   - 클래스 사용할 때: `철수.print_info()`
   - 클래스를 사용하면 마치 실제 학생에게 명령하는 것처럼 코드를 작성할 수 있어요

4. **실수를 줄일 수 있어요**:
   - 함수만 사용할 때는 매개변수의 순서를 헷갈릴 수 있어요
   - 클래스를 사용하면 각 학생의 정보가 하나로 묶여있어서 실수할 일이 줄어들어요
