## Q. 왜 정수, 실수, 문자열이 수정 불가능하죠?? 헷갈려요!
## A. 지금까지 저희는 변수에 값을 "저장"해두는 식으로 사용했어요. 자료형은 이 "값"의 종류를 뜻하는 단어에요.
## 변수를 상자에 비유한다면, 상자안에 들어가는 값의 자료형은 내용물로 볼 수 있어요.

````python
# Mutable 예제: 리스트와 딕셔너리 (수정 가능)
print("=== Mutable 예제 ===")
# 리스트는 mutable입니다.
fruits = ["사과", "바나나", "체리"]
print("원래 리스트:", fruits,"id:",id(fruits))
fruits.append("오렌지")  # 리스트에 항목 추가
print("수정된 리스트:", fruits,"id:",id(fruits))
print()

# 딕셔너리도 mutable입니다.
person = {"이름": "홍길동", "나이": 30}
print("원래 딕셔너리:", person,"id:",id(person))
person["주소"] = "서울"  # 새로운 키-값 추가
print("수정된 딕셔너리:", person,"id:",id(person))
print()

# set도 mutable입니다.
my_set = {1, 2, 3}
print("원래 set:", my_set,"id:",id(my_set))
my_set.add(4)  # set에 항목 추가
print("수정된 set:", my_set,"id:",id(my_set))
print()

print("\n=== Immutable 예제 ===")
# immutable 예제: 튜플과 문자열 (수정 불가능)
# 튜플은 immutable입니다.
coords = (10, 20, 30)
print("원래 튜플:", coords,"id:",id(coords))
try:
    coords[0] = 100  # 튜플의 값을 변경하려고 시도하면 에러가 발생합니다.
except TypeError as e:
    print("튜플 수정 시도 에러:", e)
print()

# 문자열도 immutable입니다.
greeting = "hello"
print("원래 문자열:", greeting,"id:",id(greeting))
try:
    greeting[0] = "H"  # 문자열의 문자를 변경하려고 시도하면 에러 발생
except TypeError as e:
    print("문자열 수정 시도 에러:", e)
````

````python
print("=== int 예제 (Immutable) ===")
a = 10
print("초기 a:", a, "id:", id(a))
a = 11 # a에 1을 더하면 새로운 정수 객체가 생성됩니다.
print("수정 후 a:", a, "id:", id(a))
print()

print("=== float 예제 (Immutable) ===")
x = 3.14
print("초기 x:", x, "id:", id(x))
x = 6.28  # x에 2를 곱하면 새로운 실수 객체가 생성됩니다.
print("수정 후 x:", x, "id:", id(x))
x = 3.14
print("수정 후 x:", x, "id:", id(x))
````

## ※ 진짜진짜진짜진짜 몰라도 되는 사실

위 셀을 실행하다보면 a의 id값은 동일한데 x의 id값이 막 바뀌는 모습을 보이는데 왜이럴까요?

CPython의 작은 정수 캐싱과 float 객체 생성 방식에 따른 차이에요.

CPython에서는 -5부터 256까지의 작은 정수를 미리 메모리에 할당해 놓고 재사용합니다.

추가로 20자 미만의 공백이 없는 문자열또한 캐싱이 된답니다.

※ 그럼 a의 id가 32가 차이나는 이유는 왜일까요?

a는 정수형으로 보통의 언어에서는 int형을 32 bytes로 저장하고있어요. id는 해당 값이 저장된 메모리 주소인데, 숫자형은 하나당 32 bytes를 차지하므로, 다음 숫자는 이전 숫자의 id보다 32가 많습니다

## 산술 연산자

- `+` : 두 수를 더한다
- `-` : 두 수를 뺀다.
- `*` : 두 수를 곱한다.
- `/` : 두 수를 나눈다.
- `%` : 두 수를 나눈 나머지
- `//` : 두 수를 나눈 몫

````python
a=10
b=3

# 덧셈
print("덧셈", end=": ")
print(a+b)

# 뺄셈
print("뺄셈", end=": ")
print(a-b)

# 곱셈
print("곱셈", end=": ")
print(a*b)

# 나눗셈
print("나눗셈", end=": ")
print(a/b)

# 나머지
print("나머지", end=": ")
print(a%b)

# 몫
print("몫", end=": ")
print(a//b)

# 거듭제곱
print("거듭제곱", end=": ")
print(a**b)
````

## 비교 연산자

- `==` : 양 변이 같을때 참(True) 출력
- `!=` : 양 변이 다를때 참(True) 출력
- `<` : 좌변이 우변보다 작을때 참(True) 출력
- `<=` : 좌변이 우변보다 작거나 같을때 참(True) 출력
- `>` : 좌변이 우변보다 클때 참(True) 출력
- `>=` : 좌변이 우변보다 크거나 같을때 참(True) 출력

````python
a=5
b=3
c=5
print("a:", a, "b:", b, "c:", c)
print()

# 같을때
print("같은가?")
print("a==b", end=": ")
print(a==b)
print("a==c", end=": ")
print(a==c)
print()

# 다를때
print("다른가?")
print("a!=b", end=": ")
print(a!=b)
print("a!=c", end=": ")
print(a!=c)
print()

# 좌변이 작을때
print("좌변이 작은가?")
print("a<b", end=": ")
print(a<b)
print("a<c", end=": ")
print(a<c)
print()

# 좌변이 작거나 같을때
print("좌변이 작거나 같은가?")
print("a<=b", end=": ")
print(a<=b)
print("a<=c", end=": ")
print(a<=c)
print()

# 좌변이 클때
print("좌변이 큰가?")
print("a>b", end=": ")
print(a>b)
print("a>c", end=": ")
print(a>c)
print()

# 좌변이 크거나 같을때
print("좌변이 크거나 같은가?")
print("a>=b", end=": ")
print(a>=b)
print("a>=c", end=": ")
print(a>=c)
````

## 논리 연산자

````python
a=True
b=False
c=True
d=False
print("a:", a, "b:", b, "c:", c, "d:", d)
print()

# and
print("a(", a, ") and b(", b, "):", end=" ", sep="")
print(a and b)
print("a(", a, ") and c(", c, "):", end=" ", sep="")
print(a and c)
print("b(", b, ") and d(", d, "):", end=" ", sep="")
print(b and d)
print()

# or
print("a(", a, ") or b(", b, "):", end=" ", sep="")
print(a or b)
print("a(", a, ") or c(", c, "):", end=" ", sep="")
print(a or c)
print("b(", b, ") or d(", d, "):", end=" ", sep="")
print(b or d)
print()

# not
print("not a(", a, "):", end=" ", sep="")
print(not a)
print("not b(", b, "):", end=" ", sep="")
print(not b)
````

### 예시로 시계는 60진법이에요. 60분이 지나면 분에서 시간이라는 단위가 추가되고, 분은 초기화되죠?
### 이런걸 진법이라고 하는데 컴퓨터는 이거를 두개의 수인 0과 1로 사용하는거에요

### 진법은 단위, 자릿수가 바뀌는것이라고 했어요. 그렇다면 숫자 2를 이진법으로 나타내면 어떻게 될까요?
### 0은 0으로 표시할수 있고, 1은 1로 표시할 수 있어요. 하지만 2를 표현하기 위해서는, 자릿수가 달라져야 하니깐
### 2는 10으로 표시가 가능해요. 하지만 숫자 10이랑 헷갈릴 수 있으니 10(2)와 같이 사용해요

### 이를 이용해서 비트 연산을 실행하는데,

- `|` : 비트 or 연산
- `&` : 비트 and 연산
- `^` : 비트 xor(두 값이 다를때 참) 연산
- `<< x` : 비트를 x만큼 왼쪽으로 옮기는 연산
- `>> x` : 비트를 x만큼 오른쪽으로 옮기는 연산


````python
def format_binary(n, width=8):
    """
    정수 n을 고정된 비트수(width)로 표현한 이진수 문자열로 변환합니다.
    
    이 함수는 n을 2^width (즉, 1 << width)로 나눈 나머지를 계산하여, 항상 0 이상
    2^width 미만의 값을 얻습니다. 이를 통해 n이 양수든 음수든 관계없이, 지정된 비트 수
    내에서 이진수 표현을 (2의 보수 형식으로) 제공합니다.
    """
    return format(n % (1 << width), f'0{width}b')

# 예제 사용
a = 10
b = 12
width = 4

print(f"a({a}) = {format_binary(a, width)}(2)")
print(f"b({b}) = {format_binary(b, width)}(2)")
````

````python
# 1. 비트 OR 연산 (|)
result_or = a | b
print("비트 OR (|):")
print("   " + format_binary(a, width))
print("|  " + format_binary(b, width))
print("   " + "-" * width)
print("=  " + format_binary(result_or, width), f"({result_or})")
print()
````

````python
# 2. 비트 AND 연산 (&)
result_and = a & b
print("비트 AND (&):")
print("   " + format_binary(a, width))
print("&  " + format_binary(b, width))
print("   " + "-" * width)
print("=  " + format_binary(result_and, width), f"({result_and})")
print();
````

````python
# 3. 비트 XOR 연산 (^)
result_xor = a ^ b
print("비트 XOR (^):")
print("   " + format_binary(a, width))
print("^  " + format_binary(b, width))
print("   " + "-" * width)
print("=  " + format_binary(result_xor, width), f"({result_xor})")
print()
````

````python

# 4. 왼쪽 시프트 (<<)
shift_val = 2
result_left = a << shift_val

# 왼쪽 시프트 시 자릿수가 늘어날 수 있음 -> 새로운 자리수를 계산합니다.
new_width = width + shift_val  
print(f"왼쪽 시프트 (<<) : a << {shift_val}")
print("   " + format_binary(a, width))
print("<< " + str(shift_val))
print("   " + "-" * new_width)
print("=  " + format_binary(result_left, new_width), f"({result_left})")
print()

# 5. 오른쪽 시프트 (>>)
result_right = b >> shift_val
print(f"오른쪽 시프트 (>>) : b >> {shift_val}")
print("   " + format_binary(b, width))
print(">> " + str(shift_val))
print("   " + "-" * width)
print("=  " + format_binary(result_right, width), f"({result_right})")
print()
````

## 문자열 관련 함수
### ※ []안에 있는건 선택사항

- `len(str)` : str의 길이를 반환
- `count(x, [a], [b])` : 문자열 내에서 부분 문자열 `x`가 몇 번 등장하는지 반환 (a부터 b번째까지)
- `index(x, [a], [b])` : 문자열 내에서 부분 문자열 `x`의 첫 번째 위치(인덱스)를 반환 (발견하지 못하면 ValueError 발생, a부터 b까지)
- `upper()` : 문자열의 모든 문자를 대문자로 변환한 새로운 문자열 반환
- `lower()` : 문자열의 모든 문자를 소문자로 변환한 새로운 문자열 반환
- `replace(x, y, [n])` : 문자열 내의 `x`를 찾아 `y`로 치환한 새로운 문자열 반환 (선택적으로 치환 횟수를 제한 가능, n개까지 치환)
- `split(x, [n])` : 문자열을 구분자(`x`)로 분할하여 리스트로 반환 (구분자를 지정하지 않으면 공백 기준, n개까지 나눔)

````python
# 문자열 관련 함수 예시

# 1. len
text = "Python is Good";
print("text = ", text)
print("len(text)", end=": ")
print(len(text))  # 출력: 30
print()

# 2. count
text = "hello, hello, Hello, hello";
print("text = ", text)
print("text.count(\"hello\")", end=": ")
print(text.count("hello"))  # 출력: 3
print()

# 3. index
text = "python programming"
print("text = ", text)
print("text.index(\"program\")", end=": ")
print(text.index("program"))  # 출력: 7
print()

# 4. upper
text = "Hello, World!"
print("text = ", text)
print("text.upper()", end=": ")
print(text.upper())  # 출력: HELLO, WORLD!
print()

# 5. lower
text = "Hello, World!"
print("text = ", text)
print("text.lower()", end=": ")
print(text.lower())  # 출력: hello, world!
print()

# 6. replace
text = "apple, banana, apple, orange"
print("text = ", text)
print("text.replace(\"apple\", \"grape\")", end=": ")
print(text.replace("apple", "grape"))
# 출력: grape, banana, grape, orange
print()

# 7. split
text = "one,two,three,four"
print("text = ", text)
print("text.split(\",\")", end=": ")
print(text.split(","))  # 출력: ['one', 'two', 'three', 'four']
````
